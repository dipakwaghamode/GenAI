```python
model="google/flan-t5-large"
```

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM,pipeline
from langchain import HuggingFacePipeline
```

```python
tokenizer1=AutoTokenizer.from_pretrained(model) # to generate tokens
model1=AutoModelForSeq2SeqLM.from_pretrained(model)
pipe=pipeline("text2text-generation",model=model1,tokenizer=tokenizer1)
```

```python
!pip install -U langchain-huggingface
```

```python
from langchain_huggingface import HuggingFacePipeline
```

```python
llm=HuggingFacePipeline(pipeline=pipe,model_kwargs={"temperature":0.5,"max_length":1000})
```

```python
from langchain.chains import RetrievalQA
```

```python
from langchain.prompts import PromptTemplate
template="""use the context to provide a concise answer and if you don't know just say don't now.
{context}
Question: {question}
Helpful Answer:"""

prompt_template=PromptTemplate.from_template(template)
```

```python
qa_chain=RetrievalQA.from_chain_type(llm,chain_type="stuff",retriever=vectordatabase.as_retriever(),chain_type_kwargs={"prompt":prompt_template})
```

```python
question="who is the CEO of company?"
result=qa_chain({"query":question})
```
